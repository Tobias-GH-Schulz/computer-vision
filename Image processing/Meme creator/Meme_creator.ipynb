# A basic Meme creator with one image and a short line of text, using any openCV font

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import ImageFont, ImageDraw, Image 
```

```python
pic = cv2.imread("./input/funny_animal.jpg")
pic = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,7))
plt.imshow(pic);
```

```python
def meme_creator(path, text):
    # loading the image and converting it to RGB
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    font = cv2.FONT_ITALIC

    # get boundary of this text
    textsize = cv2.getTextSize(text, font, 1, 2)[0]

    # get coords based on boundary
    textX = int((img.shape[1] - textsize[0]) / 2)
    textY = int(img.shape[0] - 85) 

    # put a black rectangle with white text on the image
    cv2.rectangle(img, (0,250), (699,300), (0,0,0), cv2.FILLED)
    cv2.putText(img, text, (textX, textY ), font, 1, (255, 255, 255), 2)

    # export the meme
    cv2.imwrite("./output/meme.jpg", cv2.cvtColor(img, cv2.COLOR_RGB2BGR))

    # show the meme
    plt.figure(figsize=(10,7))
    plt.imshow(img)
```

```python
meme_creator("./input/funny_animal.jpg", "Have a nice day!")
```
